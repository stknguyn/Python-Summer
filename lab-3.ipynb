```python
class Person:
    count = 0

    def __init__(self, name, age):
        self.__name = name
        self.__age = age
        Person.increment_count()
# method getter setter

    @property
    def name(self):
        return self.__name

    @name.setter
    def name(self, name):
        self.__name = name

    @property
    def age(self):
        return self.__age

    @age.setter
    def age(self, age):
        self.__age = age
# another method

    @classmethod
    def increment_count(cls):
        cls.count += 1

    def greeting(self):
        print("Hello")

    @staticmethod
    def cls_information():
        print(f"The information of class: ")
        print(f"\t Class name: {Person.__name__}")
        print(f"\t Base class: {Person.__base__}")
        print(f"\t The numbers of object: {Person.count}")

    def __str__(self) -> str:
        str = ''
        str += '='*40
        str += f"\nHello my name is {self.__name}, I'm {self.__age} years old\n" 
        str += '='*40
        return str
```

```python
class Manager(Person):
    man_count = 0

    def __init__(self, name, age, salary, bonus):
        super().__init__(name, age)
        self.__salary = salary
        self.__bonus = bonus
        Manager.increment_count()

    @property
    def salary(self):
        return self.__salary

    @salary.setter
    def salary(self, salary):
        self.__salary = salary

    @property
    def bonus(self):
        return self.__bonus

    @bonus.setter
    def bonus(self, bonus):
        self.__bonus = bonus

    @classmethod
    def increment_count(cls):
        cls.man_count += 1

    @staticmethod
    def cls_information():
        print(f"The information of Manager class: ")
        print(f"\t Class name: {Manager.__name__}")
        print(f"\t Base class: {Manager.__base__.__name__}")
        print(f"\t The numbers of Person objects: {Person.count}")
        print(f"\t The numbers of Manager objects: {Manager.man_count}")

    def __str__(self) -> str:
        str_repr = super().__str__()
        str_repr += f"\nSalary: {self.__salary}, Bonus: {self.__bonus}\n"
        str_repr += '='*40
        return str_repr

# Test the classes
p1 = Person("Alice", 30)
p2 = Person("Bob", 25)
m1 = Manager("Charlie", 40, 100000, 20000)

print(p1)
print(p2)
print(m1)

Person.cls_information()
Manager.cls_information()
```
