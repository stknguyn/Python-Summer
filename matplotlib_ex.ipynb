```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
x_value = np.linspace(-10,10,200)
y_value = np.sin(x_value)
plt.plot(x_value,y_value,linestyle='-.',color='green')
plt.show()
```

```python
plt.plot(x_value,y_value,linestyle='',color='green',marker='o')
```

```python
#plot only number from the index 50 to 80
plt.plot(x_value[50:80],y_value[50:80],linestyle='--',color='green',marker='o')
plt.show()
plt.plot(x_value[51:56],y_value[51:56],linestyle='--',color='blue',marker='x')
plt.show()
```

```python
with open('weight-height.csv') as f:
    data = pd.read_csv(f)
print(data)
data_vl = data.values
#plot the height and weight
plt.plot(data_vl[:,1],data_vl[:,2],linestyle='',color='blue',marker='o')
```

```python
#plot the male with blue, female with red in the same plot
# Plot the male with blue, female with red in the same plot
male_data = data[data['Gender'] == 'Male']
female_data = data[data['Gender'] == 'Female']

plt.plot(male_data['Height'], male_data['Weight'], 'bo', label='Male')
plt.plot(female_data['Height'], female_data['Weight'], 'ro', label='Female')

plt.xlabel('Height')
plt.ylabel('Weight')
plt.legend()
plt.show()
```

```python
with open('company-sales.csv') as f:
    data = pd.read_csv(f)
months = data['month_number'].values
total_profit = data['total_profit'].values
plt.plot(months, total_profit)
plt.xlabel('Month Number')
plt.xticks(months)
plt.ylabel('Total Profit')
plt.title('Company profit per month')
plt.show()
```

```python
plt.plot(months, total_profit, linestyle='--', color='red', marker='o',markerfacecolor='black')
plt.legend(['Profit data of last year'],loc='lower right')
plt.xlabel('Month Number')
plt.ylabel('Profit in dollar')
plt.xticks(months)
plt.title('Company Sales data of last year')
plt.show()
```

```python
#read all product sales, and plot the sales of each product in the same plot
with open('company-sales.csv') as f:
    data = pd.read_csv(f)
months = data['month_number'].values
facecream = data['facecream'].values
facewash = data['facewash'].values
toothpaste = data['toothpaste'].values
bathingsoap = data['bathingsoap'].values
shampoo = data['shampoo'].values
moisturizer = data['moisturizer'].values
plt.plot(months, facecream, 'o-', )
plt.plot(months, facewash, 'o-',)
plt.plot(months, toothpaste, 'o-', )
plt.plot(months, bathingsoap, 'o-', )
plt.plot(months, shampoo, 'o-', )
plt.plot(months, moisturizer, 'o-', )
plt.legend(['Face cream', 'Face wash', 'Tooth paste', 'Bathing soap', 'Shampoo', 'Moisturizer'])
```

```python
df = pd.read_csv("company-sales.csv")
monthList = df ['month_number'].values
faceCremSalesData = df ['facecream'].values
faceWashSalesData = df ['facewash'].values
plt.bar([a-0.25 for a in monthList], faceCremSalesData, width= 0.25, label ='Face Cream sales data', align='edge')
plt.bar([a+0.25 for a in monthList], faceWashSalesData, width= -0.25, label ='Face Wash sales data', align='edge')
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.legend(loc='upper left')
plt.title('Sales data')
plt.xticks(monthList)
plt.grid(True, linewidth= 1, linestyle="--")
plt.title('Facewash and facecream sales data')
plt.show()
```

```python
total_profit = df['total_profit'].values
print(total_profit)
plt.hist(total_profit, bins=10, edgecolor='black')
plt.xlabel('Profit range in dollar')
plt.ylabel('Actual Profit in dollar')
plt.title('Profit data')
plt.legend(['Profit data'],loc='upper left')
```

```python
# Read Bathing soap and facewash of all months and display it using the Subplot
plt.subplot(1,2,2)
plt.plot(monthList, bathingsoap, 'o-', label='Bathing Soap Sales Data')
plt.plot(monthList, facewash, 'o-', label='Facewash Sales Data')
plt.title('Bathing Soap and Facewash Sales Data')
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.legend()
plt.tight_layout()
plt.show()
```
